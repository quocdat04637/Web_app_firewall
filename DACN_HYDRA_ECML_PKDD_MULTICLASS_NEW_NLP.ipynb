```python
from urllib.parse import unquote

def decode_text(encoded_text):
    try:
        decoded_text = unquote(encoded_text)
        return decoded_text
    except UnicodeDecodeError:
        print("Unable to decode the text.")
        return None
```

```python
import re
def separate_strings(s):
    # Thêm khoảng trắng trước và sau ký tự đặc biệt
    s = re.sub(r'(\W)', r' \1 ', s)
    # Tách chuỗi dựa trên khoảng trắng
    result = s.split()
    return result
```

```python
def find_common_elements(list1, list2, list3):
    set1 = set(list1)
    set2 = set(list2)
    set3 = set(list3)
    common_elements = set1.intersection(set2, set3)
    return list(common_elements)
```

```python
for index in range(20):
  # Tìm các vị trí giống nhau của X_train_model1[index] trong padded_sequences1
  matching_indices1 = np.where(np.all(padded_sequences1 == X_train_model1[index], axis=1))[0]

  # Tìm các vị trí giống nhau của X_train_model2[index] trong padded_sequences2
  matching_indices2 = np.where(np.all(padded_sequences2 == X_train_model2[index], axis=1))[0]

  # Tìm các vị trí giống nhau của X_train_model3[index] trong padded_sequences3
  matching_indices3 = np.where(np.all(padded_sequences3 == X_train_model3[index], axis=1))[0]

  match_index = find_common_elements(matching_indices1, matching_indices2, matching_indices3)

  print("X_train_model1[{}], X_train_model2[{}], X_train_model3[{}] are created from padded_sequences1{}, padded_sequences2{}, padded_sequences3{}".format(index, index, index, match_index, match_index, match_index))
```

# Xây dựng multi-modal hydra

1. Model 1

```python
# Xây dựng mô hình Model 1
model1 = Sequential()
model1.add(Embedding(4000, 4, input_length=130)) # embedding layer
model1.add(Flatten())
model1.add(Dense(250)) # fully connected layer
```

```python
model1.summary()
```

2. Model 2

```python
# Xây dựng mô hình Model 2
model2 = Sequential()
model2.add(Embedding(4000, 8, input_length=250)) # embedding layer
model2.add(Conv1D(48, 32, strides=1))
model2.add(Conv1D(96, 32, strides=1))
model2.add(MaxPooling1D(4))  # Thêm lớp MaxPooling1D
model2.add(Conv1D(128, 16, strides=2))
model2.add(Conv1D(192, 16, strides=2))
model2.add(GlobalAveragePooling1D())  # Thêm lớp GlobalAveragePooling1D
model2.add(Dense(192)) # fully connected layer
model2.add(Dense(160)) # fully connected layer
model2.add(Dense(128)) # fully connected layer
```

```python
model2.summary()
```

```python
# Kết nối 2 lớp FC cuối cùng của 2 model
combined_fc12 = Concatenate()([model1.layers[-1].output, model2.layers[-1].output])
fc_layer12 = Dense(150)(combined_fc12)
```

```python
fc_layer12.shape
```

```python
# Xây dựng mô hình Model 3
# Kích thước embedding
embedding_size = 4
# Kích thước các bộ lọc
filter_sizes = [3, 4, 5]
# Số lượng bộ lọc
num_filters = 300

model3 = Sequential()
# Lớp Embedding
model3.add(Embedding(4000, embedding_size, input_length=250))
# Lớp Convolutional và Activation
for filter_size in filter_sizes:
    model3.add(Conv1D(num_filters, filter_size, activation='elu'))
# Lớp Global Max-Pooling
model3.add(GlobalMaxPooling1D())

# Lấy output từ lớp GlobalMaxPooling1D
output3 = model3.layers[-1].output
```

```python
model3.summary()
```

```python
output3.shape
```

```python
# Kết nối fc_layer12 với output3
combined_output = Concatenate()([fc_layer12, output3])

# Thêm lớp FC(200)
fc = Dense(200)(combined_output)
```

```python
fc.shape
```

```python
# Tạo lớp Output layer
output_layer = Dense(8, activation='softmax')(fc)
```

```python
output_layer.shape
```

Multi classification

```python
from tensorflow import keras
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Concatenate

# Tạo multi-modal model
multi_modal_model = Model(inputs=[model1.input, model2.input, model3.input], outputs=output_layer)
multi_modal_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics='accuracy')
multi_modal_model.summary()
# Huấn luyện multi-modal model
history = multi_modal_model.fit([X_train_model1, X_train_model2, X_train_model3], k_train, epochs=5)
```

```python
import matplotlib.pyplot as plt

# Lấy các giá trị loss và accuracy từ biến history
loss = history.history['loss']
accuracy = history.history['accuracy']

# Vẽ đồ thị loss và accuracy trên cùng một đối tượng đồ thị
plt.plot(loss, label='Loss')
plt.plot(accuracy, label='Accuracy')

plt.title('Training Loss and Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
normal_attack_lists = [
    "Normal",
    "LdapInjection",
    "OsCommanding",
    "PathTransversal",
    "SSI",
    "SqlInjection",
    "XPathInjection",
    "XSS"]
```

```python
# Tính toán precision, recall, và f1-score
report = classification_report(k_test.values.argmax(axis=1), k_pred_test.argmax(axis=1), digits=4, output_dict=True)
print(classification_report(k_test.values.argmax(axis=1), k_pred_test.argmax(axis=1), digits=4))
# Chuyển đổi dữ liệu thành DataFrame
df = pd.DataFrame(report).transpose()
# Lưu DataFrame vào file CSV
df.to_csv('/content/drive/MyDrive/NT114_DACN/ECML_PKDD4/classification_report_multi_modal_multiclass.csv', index=True)

# Compute the confusion matrix
cm = confusion_matrix(k_test.values.argmax(axis=1), k_pred_test.argmax(axis=1))
# Lưu confusion matrix vào DataFrame
df = pd.DataFrame(cm)
# Lưu DataFrame vào file CSV
df.to_csv('/content/drive/MyDrive/NT114_DACN/ECML_PKDD4/confusion_matrix_multi_modal_multiclass.csv', index=False)


# Create a heatmap using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', annot_kws={"ha": 'center', "va": 'center'})

# Set tick labels
tick_labels = [normal_attack_lists[i] for i in range(len(normal_attack_lists))]
plt.xticks(ticks=np.arange(len(normal_attack_lists)) + 0.5, labels=tick_labels, rotation=90)
plt.yticks(ticks=np.arange(len(normal_attack_lists)) + 0.5, labels=tick_labels, rotation=0)

plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
import numpy as np

# Mảng nhãn thực tế và mảng xác suất dự đoán
true_labels = k_test.values.argmax(axis=1)
probs = k_pred_test

# Chuyển đổi nhãn thành dạng nhị phân
n_classes = len(np.unique(true_labels))
binarized_labels = label_binarize(true_labels, classes=np.arange(n_classes))

# Tính toán ROC curve và AUC cho từng lớp
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(binarized_labels[:, i], probs[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Tính toán ROC curve và AUC tổng thể
macro_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))
macro_tpr = np.zeros_like(macro_fpr)
for i in range(n_classes):
    macro_tpr += np.interp(macro_fpr, fpr[i], tpr[i])
macro_tpr /= n_classes
macro_auc = auc(macro_fpr, macro_tpr)

# In kết quả
print("Macro Average:")
print("FPR:", np.round(macro_fpr, decimals=4))
print("TPR:", np.round(macro_tpr, decimals=4))
print("AUC:", np.round(macro_auc, decimals=4))

# Trực quan hóa ROC curve cho từng lớp
plt.figure(figsize=(8, 6))
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], label='%s (AUC = %0.4f)' % (normal_attack_lists[i], roc_auc[i]))

# Trực quan hóa ROC curve tổng thể
plt.plot(macro_fpr, macro_tpr, label='Macro Average (AUC = %0.4f)' % macro_auc, color='black', linestyle='--')

# Định dạng đồ thị
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.show()
```
